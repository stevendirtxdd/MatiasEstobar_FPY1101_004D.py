```python
productos = {
    '8475HD': ['HP', 15.6, '8GB', 'DD', '1T', 'Intel Core i5', 'Nvidia GTX1050'],
    '2175HD': ['Lenovo', 14, '4GB', 'SSD', '512GB', 'Intel Core i5', 'Nvidia GTX1050'],
    'JjfFHD': ['Asus', 14, '16GB', 'SSD', '256GB', 'Intel Core i7', 'Nvidia RTX2080Ti'],
    'fgdxFHD': ['HP', 15.6, '8GB', 'DD', '1T', 'Intel Core i3', 'integrada'],
    'GF75HD': ['Asus', 15.6, '8GB', 'DD', '1T', 'Intel Core i7', 'Nvidia GTX1050'],
    '123FHD': ['Lenovo', 14, '6GB', 'DD', '1T', 'AMD Ryzen 5', 'integrada'],
    '342FHD': ['Lenovo', 15.6, '8GB', 'DD', '1T', 'AMD Ryzen 7', 'Nvidia GTX1050'],
    'UWU131HD': ['Dell', 15.6, '8GB', 'DD', '1T', 'AMD Ryzen 3', 'Nvidia GTX1050']
}

stock = {
    '8475HD': [387990, 10], '2175HD': [327990, 4], 'JjfFHD': [424990, 1],
    'fgdxFHD': [664990, 21], '123FHD': [290890, 32], '342FHD': [444990, 7],
    'GF75HD': [749990, 2], 'UWU131HD': [349990, 1], 'FS1230HD': [249990, 0]
}

def stock_marca(marca: str) -> None:
    total = sum(stock.get(modelo, [0, 0])[1]
                for modelo, datos in productos.items()
                if datos[0].casefold() == marca.casefold())
    print(f"El stock total de la marca {marca.title()} es: {total}")

def buscar_por_precio(p_min: str, p_max: str) -> None:
    if not (p_min.isdigit() and p_max.isdigit()):
        print("¡¡Debe ingresar valores enteros!!")
        return

    p_min, p_max = int(p_min), int(p_max)
    disponibles = [
        f"{productos[modelo][0]}--{modelo}"
        for modelo, (precio, cantidad) in stock.items()
        if p_min <= precio <= p_max and cantidad > 0 and modelo in productos
    ]

    if disponibles:
        print("Productos disponibles:", sorted(disponibles))
    else:
        print("No hay productos en ese rango de precio.")

def actualizar_precio(modelo: str, nuevo_precio: int) -> bool:
    if modelo in stock:
        stock[modelo][0] = nuevo_precio
        return True
    return False

def menu() -> None:
    while True:
        print("\n*** MENÚ PRINCIPAL ***")
        print("1. Stock por marca")
        print("2. Buscar por rango de precios")
        print("3. Actualizar precio de modelo")
        print("4. Salir")
        opcion = input("Ingrese opción: ").strip()

        if opcion == '1':
            marca = input("Ingrese marca a consultar: ").strip()
            stock_marca(marca)

        elif opcion == '2':
            p_min = input("Ingrese precio mínimo: ").strip()
            p_max = input("Ingrese precio máximo: ").strip()
            buscar_por_precio(p_min, p_max)

        elif opcion == '3':
            while True:
                modelo = input("Ingrese modelo a actualizar: ").strip()
                if modelo not in stock:
                    print("¡¡El modelo no existe!!")
                    continue

                nuevo_precio_input = input("Ingrese precio nuevo: ").strip()
                if not nuevo_precio_input.isdigit():
                    print("¡Debe ingresar un precio válido (entero)!")
                    continue

                nuevo_precio = int(nuevo_precio_input)
                if actualizar_precio(modelo, nuevo_precio):
                    print("¡¡Precio actualizado!!")

                seguir = input("¿Desea actualizar otro precio (s/n)?: ").strip().lower()
                if seguir != 's':
                    break

        elif opcion == '4':
            print("Cerrando programa....")
            break

        else:
            print("¡¡Debe seleccionar una opción válida!!")

menu()

```
